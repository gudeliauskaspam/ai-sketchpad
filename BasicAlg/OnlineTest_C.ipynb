```python
def find_region(grid, row, col, idx):
    assert idx != 0
    if row < 0 or row >= len(grid) or col < 0 or col >= len(grid[0]):
        return 0
    elif grid[row][col] != idx:
        return 0
    
    grid[row][col] = 0
    size = 1
    
    size += find_region(grid, row-1, col, idx)
    size += find_region(grid, row+1, col, idx)
    size += find_region(grid, row, col+1, idx)
    size += find_region(grid, row, col-1, idx)
    
    return size

def find_nb_regions(grid):
    
    nb_regions = 0
    largest_reg = 0
    
    for row in range(0, len(grid)):
        for col in range(0, len(grid[0])):
            idx = grid[row][col]
            if idx != 0:
                nb_regions += 1
                size = find_region(grid, row, col, idx)
                if size > largest_reg:
                    largest_reg = size
                
    return nb_regions, largest_reg
```

```python
n = 4
m = 4
grid = [[1, 1, 0, 0],
        [0, 1, 1, 0],
        [0, 0, 1, 0],
        [1, 0, 0, 0]]
```

```python
nb, largest = find_nb_regions(grid)
print(nb, largest)
```
